```python
# b)
def two_dimensional_gaussian(mean, cov, pos):
    cov_det = np.linalg.det(cov)
    cov_inv = np.linalg.inv(cov)
    
    N = np.sqrt((2 * np.pi) ** 2 * cov_det)
    
    fac = np.einsum('...k,kl,...l->...', pos - mean, cov_inv, pos - mean)

    return np.exp(-fac / 2) / N

sample = 100


X = np.linspace(-3, 3, sample)
Y = np.linspace(-2, 4, sample)
X, Y = np.meshgrid(X, Y)
```

```python
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

mean = np.array([0, 1])
cov = np.array([[1, 0.25], [0.25, 1]])

dst = two_dimensional_gaussian(mean, cov, pos)

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y, dst, linewidth=1, cmap=cm.jet)

ax.set_zlim(-0.05, 0.2)
ax.set_zticks(np.linspace(0, 0.2, 5))
ax.view_init(20, -20)

plt.show()
```

```python
# The result can be checked against the multivariate gaussian method in SciPy
dst = multivariate_normal(mean, cov)

pdf = np.zeros(X.shape)

for (i, _) in enumerate(X):
    for (j, _) in enumerate(Y):
        pdf[i, j] = dst.pdf([X[i, j], Y[i, j]])

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y, pdf, linewidth=1, cmap=cm.jet)

ax.set_zlim(-0.05, 0.2)
ax.set_zticks(np.linspace(0, 0.2, 5))
ax.view_init(20, -20)

plt.show()
```
