```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

from scipy.linalg import cholesky, cho_solve, cho_factor
from scipy.spatial.distance import pdist, cdist, squareform
from sklearn.metrics.pairwise import rbf_kernel
```

```python
# Gaussian Kernel
kernel_width = 1

def gaussian_kernel_dist(dist):
    return np.exp(-dist / kernel_width)

def gaussian_kernel(x, xp):
    return np.exp(-np.linalg.norm(x, xp) / kernel_width)
```

```python
X_train = np.array([
    [-2],
    [0],
    [1]
])

y_train = np.array([
    2,
    0,
    2
])
```

```python
distances = pdist(X_train, 'sqeuclidean')
print('Distances between points:', distances)

print('Gaussian kernel applied to distances:')
print(gaussian_kernel_dist(distances))

K = squareform(gaussian_kernel_dist(distances)) + np.identity(X_train.shape[0])

print('Full gaussian kernel:')
print(K)
```

```python
c, low = cho_factor(K)

print(c)
print(low)
```

```python
cho_solve((c, low), y_train)
```

```python
np.linalg.inv(K) @ y_train
```

```python
print('scikit-learn rbf_kernel:')
rbf_kernel(X_train)
```

```python
class GaussianKernelRegressor:
    
    def __init__(self, kernel_width=1):
        self.kernel_width = kernel_width
    
    def _gaussian_kernel_distances(self, distances):
        return np.exp(-distances / self.kernel_width)
    
    def _gaussian_kernel(self, x, xp):
        return np.exp(-np.linalg.norm(x, xp) / self.kernel_width)
        
    def fit(self, X_train, y_train):
        if X_train is None:
            raise Exception('Training data input has not been provided')
            
        self.X_train = X_train
            
        if y_train is None:
            raise Exception('Training data output has not been provided')
            
        distances = sp.spatial.distance.pdist(X_train, 'sqeuclidean')
        
        A = sp.spatial.distance.squareform(self._gaussian_kernel_distances(distances)) + np.identity(self.X_train.shape[0])
        
        A_factor = sp.linalg.cho_factor(A)
        
        self.coef_ = sp.linalg.cho_solve(A_factor, y_train)
        
        return self
    
    def predict(self, X_eval):
        distances = sp.spatial.distance.cdist(X_eval, self.X_train, 'sqeuclidean')
        
        kernels = self._gaussian_kernel_distances(distances)
        
        return np.einsum('ij,kj->ki', [self.coef_], kernels)
```

```python
X_eval = np.linspace(-3, 3, 600).reshape(-1, 1)
```

```python
gkr = GaussianKernelRegressor()
gkr.fit(X_train, y_train)
gkr_pred = gkr.predict(X_eval)
```

```python
# Plot model and the training points
plt.figure(figsize=(7, 7))

X_eval = np.linspace(-3, 3, 600).reshape(-1, 1)

plt.plot(X_eval, gkr.predict(X_eval), color='g', label='Gaussian regression model')
plt.plot(X_train, y_train, 'r.', markersize=10, label='Training set')

plt.legend()
plt.show()
```

```python
class KernelRidgeRegressor:
    
    def __init__(self, regularization=1, kernel_width=1):
        self.regularization = regularization
        self.kernel_width = kernel_width
    
    def _gaussian_kernel_distances(self, distances):
        return np.exp(-distances / self.kernel_width)
    
    def _gaussian_kernel(self, x, xp):
        return np.exp(-np.linalg.norm(x, xp) / self.kernel_width)
        
    def fit(self, X_train, y_train):
        if X_train is None:
            raise Exception('Training data input has not been provided')
            
        self.X_train = X_train
            
        if y_train is None:
            raise Exception('Training data output has not been provided')
            
        distances = sp.spatial.distance.pdist(X_train, 'sqeuclidean')
        
        A = sp.spatial.distance.squareform(self._gaussian_kernel_distances(distances)) + np.identity(self.X_train.shape[0])
        
        A_factor = sp.linalg.cho_factor(A + self.regularization * np.identity(self.X_train.shape[0]))
        
        self.coef_ = sp.linalg.cho_solve(A_factor, y_train)
        
        return self
    
    def predict(self, X_eval):
        distances = sp.spatial.distance.cdist(X_eval, self.X_train, 'sqeuclidean')
        
        kernels = self._gaussian_kernel_distances(distances)
        
        return np.einsum('ij,kj->ki', [self.coef_], kernels)
```

```python
krr = KernelRidgeRegressor()
krr.fit(X_train, y_train)
krr_pred = krr.predict(X_eval)
```

```python
# Plot model and the training points
plt.figure(figsize=(7, 7))

X_eval = np.linspace(-3, 3, 600).reshape(-1, 1)

plt.plot(X_eval, krr.predict(X_eval), color='g', label='Gaussian regression model')
plt.plot(X_train, y_train, 'r.', markersize=10, label='Training set')

plt.legend()
plt.show()
```

```python
# Plot both models and the training points
plt.figure(figsize=(7, 7))

X_eval = np.linspace(-3, 3, 600).reshape(-1, 1)

plt.plot(X_eval, gkr.predict(X_eval), color='g', label='Gaussian regression model with kernel width = 1')
plt.plot(X_eval, krr.predict(X_eval), color='b', label='Kernel ridge regression model with $\lambda = 1$')
plt.plot(X_train, y_train, 'r.', markersize=10, label='Training set')

plt.legend(loc='upper right')
plt.ylim([-1, 3])
plt.savefig('p02.png', dpi=500)
plt.show()
```
