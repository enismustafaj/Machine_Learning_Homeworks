```python
import matplotlib.pyplot as plt
import numpy as np
import random

np.random.seed(0)
```

```python
no_points = 40

irr_error_var = 50
variance = 2 ** np.linspace(0, 10, no_points)
bias_squared = 2 ** np.linspace(10, 0, no_points)

total_error = [var + bias_sq + irr_error_var for var, bias_sq in zip(variance, bias_squared)]
x_axis_ticks = np.arange(variance.shape[0])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(x_axis_ticks, variance, 'g-', label='Variance')
ax.plot(x_axis_ticks, bias_squared, 'r-', label='Squared bias')
ax.plot(x_axis_ticks, total_error, 'b:', label='Total error')
ax.axhline(y=irr_error_var, color='r', linestyle=':', label='Irreducible error')

ax.legend(loc='upper center')
ax.set_xlabel('Model complexity')

ax.set_title('Bias-Variance Tradeoff')

plt.savefig('bias_variance_plot.png', dpi=300)
plt.show()
```
